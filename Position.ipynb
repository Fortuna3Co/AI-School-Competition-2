```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('c:\\Users\\hskang\\Desktop\\Coding\\Python AI\\Datas\\train.csv')
test = pd.read_csv('c:\\Users\\hskang\\Desktop\\Coding\\Python AI\\Datas\\test.csv')
bus_bts = pd.read_csv('c:\\Users\\hskang\\Desktop\\Coding\\Python AI\\Datas\\bus_bts.csv')

# 공공 데이터 포털 https://www.data.go.kr/data/15010850/fileData.do
# 데이터 기준 일자 2017-08-31
station_position = pd.read_csv('c:\\Users\\hskang\\Desktop\\Coding\\Python AI\\Datas\\제주특별자치도_버스정류소현황_20170831.csv', engine='python', encoding='cp949')
```

```python
get_station_position = pd.DataFrame()
get_station_position['station_name'] = station_position['정류소명']
get_station_position['latitude'] = station_position['위도']
get_station_position['longitude'] = station_position['경도']
get_station_position = get_station_position.drop_duplicates(['station_name']).reset_index(drop=True)
get_station_position
```

```python
train = pd.merge(train, get_station_position, on='station_name', how='left')
test = pd.merge(test, get_station_position, on='station_name', how='left')
```

```python
train_null_position = train[train['longitude'].isnull()]
test_null_position = test[test['longitude'].isnull()]
```

```python
null_position = pd.concat([train_null_position['station_name'], test_null_position['station_name']]).unique()
null_position
```

```python
print(len(null_position))
```

```python
import urllib
import re
import requests
import json
from urllib.parse import quote
from urllib.request import urlopen

#제주 데이터 허브 이용

your_projectKey = 'c0p07bb7j___j_b_r8722br9o8tbot__'
get_empty_station_position = {'station_name':[], 'longitude':[], 'latitude':[]}

for position in null_position:
    get_empty_station_position['station_name'].append(position)
    api_url = 'https://open.jejudatahub.net/api/proxy/DD11ab6a6t11D16baaa1a2tD26ata161/' + your_projectKey + '?{stationName(key=' + quote(position) + ')}'
    data = json.loads(urlopen(api_url).read())

    get_empty_station_position['longitude'].append(data['data'][0]['longitude'])
    get_empty_station_position['latitude'].append(data['data'][0]['latitude'])
```

```python
pd.DataFrame(get_empty_station_position)
```

```python
all_station_position = pd.concat([get_station_position, pd.DataFrame(get_empty_station_position)])
```
